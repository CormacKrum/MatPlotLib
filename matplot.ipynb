# Plotting the ball's Trajectory


```python
import numpy as np
import matplotlib.pyplot as plt
h=1.5
g=9.8
v=4
angle=45
```

```python
t=np.linspace(0,1,1000)
```

```python
x=v*np.cos(angle*np.pi/180)*t
y=h+v*np.sin(angle*np.pi/180)*t-g*t*t/2
```

```python
coun=0
for i in y:
    if i<0:
        break
    else:
        coun+=1
print("The Ball hits the ground at the distance of {} meters".format(x[coun]))
```

```python
plt.figure()
plt.plot(x,y,label="Ball")
plt.xlim(0)
plt.ylim(0)
plt.title("Ball Trajectory")
plt.xlabel("Horizontal Distance")
plt.ylabel("Height")
plt.legend()
plt.show()
```

# Convergence of Infinite Series

```python
p=0.99
k=np.arange(1001)
```

```python
p_k=p**k
```

```python
G=1/(1-p)
glist=[]
for i in range(1001):
    glist.append(G)
```

```python
plt.figure()
plt.plot(k,glist,color="red",label="Infinity Sum")
plt.plot(k,np.cumsum(p_k),color="blue",label="Finite Sum")
plt.title("Geometric Series")
plt.xlabel("k")
plt.ylabel("Cumulative sum of Geom Series")
plt.legend()
plt.show()
```

# Lets Try this Thing with Another Series

```python
p=2
n=np.arange(1,501)
```

```python
pSeries=1/(n**p)
```

```python
P=np.pi**2/6
Pl=[]
for i in range(500):
    Pl.append(P)
```

```python
plt.figure()
plt.plot(n,Pl,label="Infinite Sum",color="red")
plt.plot(n,np.cumsum(pSeries),label="Finite Sum",color="blue")
plt.legend()
plt.title("Sum Of Series")
plt.xlabel("n")
plt.ylabel("Cummlative sum of Series ")
plt.show()
```
