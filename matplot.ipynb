```python
center=np.array([50,50])
#max_distance=50*np.sqrt(2)
for i in range(100):
    for j in range(100):
        x=np.sqrt(((center-np.array([i,j]))**2).sum())
        #color=255-x/max_distance*255
        color=255-(int(x)*1000)%256
        img[i,j,0]=color

plt.imshow(img)
```

```python
center=np.array([img.shape[0]/2,img.shape[1]/2])

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        x=np.sqrt(((center-np.array([i,j]))**2).sum())
        #color=255-x/max_distance*255
        color=255-(int(x)*100)%256
        img[i,j,0:2]=color

plt.imshow(img)
```

```python
import math
center=np.array([50,50])
max_distance=50*np.sqrt(2)
for i in range(100):
    for j in range(100):
        x=np.sqrt(((center-np.array([i,j]))**2).sum())
        color=255-x/max_distance*255
        color=255-x/max_distance*255
        w=(int(color)//int(max_distance))%3
        img[i,j,w]=color

plt.imshow(img)
```

```python
img=np.zeros((100,100,3),dtype=np.uint8)
plt.imshow(img)
center=np.array([50,50])
max_distance=50*np.sqrt(2)
for i in range(100):
    for j in range(100):
        x=np.sqrt(((center-np.array([i,j]))**2).sum())
        color=255-x/max_distance*255
        #color=255-(int(x)*1000)%256
        img[i,j,0]=color

plt.imshow(img)
```
